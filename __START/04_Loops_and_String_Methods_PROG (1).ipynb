```python
# 0. Import modules:
import pprint as pp
import random
import numpy as np
```

```python
# 1. Define a range(start,stop):
nums_list = list(range(1,11))
print(nums_list)
```

```python
# 2. Loop the range of consec ints from 1 to 10 (exclusive):
# n (or some other variable name) represents the current item
# print n squared each time the loop runs
# when it reaches the end of the range, the loop stops
for n in nums_list:
  print(n**2)
```

```python
# 3. CHALLENGE: Loop the nums list (which was made from a range):
# if n is even, square it (n ** 2)
# else, n is odd, so cube it (n ** 3)
# print the result each time from 1-10
# expected result: 1 4 9 16 125 36 ... 100
# HINT: %
for n in nums_list:
  if n % 2 == 1: # if odd
    print(n**3)
  else:
    print(n**2)

print()
# BONUS: make a new list of results and then print that at the end
squares_and_cubes = []

for n in nums_list:
  if n % 2 == 1: # if odd
    squares_and_cubes.append(n**3)
  else:
    squares_and_cubes.append(n**2)

print(squares_and_cubes)
# [1, 4, 27, 16, 125, 36, 343, 64, 729, 100]

# get the sum, mean, min, max, count
sum_tot = sum(squares_and_cubes)
print(sum_tot)

mean_value = np.mean(squares_and_cubes)
print(mean_value)

# get min max
min_num = min(squares_and_cubes)
print(min_num)

max_num = max(squares_and_cubes)
print(max_num)

# count num of occurrences of some value in list
nums = [1, 4, 4, 27, 16, 4, 125, 36, 343, 4, 64, 729, 4, 100, 4, 4]
num_4s = nums.count(4)
print(num_4s)
```

```python
# 4. Challenge: given this list of fruits:
fruits = ['apple','banana','blueberry','cherry','grape','kiwi','lemon','lime','mango', 'orange','peach','pineapple','raspberry','strawberry','tangerine','watermelon']

# challenge: do a fruit loop and make jellybeans: "apple jellybean" etc.
# save each to a new list called jellybeans and print the list:
# ['apple jellybean', 'banana jellybean'.. etc. .. 'watermelon jellybean']
jellybeans = []
# do a fruit loop
for fru in fruits:
  jellybeans.append(fru + " jellybean")

pp.pprint(jellybeans[::2])
```

```python
# i.) loop the fruits list, making jellybeans of everything EXCEPT
#     fruits that start with "p" -- make popsicles out of those
#     save results to new list called treats
#.    but first add 5 more p fruits and then shuffle the list
p_fruits = ['plum','papaya','pomegranate','pear','persimmon']
fruits.extend(p_fruits)
random.shuffle(fruits)

jbeans_or_popsicles = []

for fru in fruits:
  if fru.startswith('p'):
    jbeans_or_popsicles.append(fru + " popsicle")
  else:
    jbeans_or_popsicles.append(fru + " jellybean")

pp.pprint(jbeans_or_popsicles)
```

```python
# FRUIT LOOP CHALLENGE:
treats = []
# loop the fruits list and do stuff:
for fru in fruits:
    fru_dict = {"fruit":fru}
    #. √ if the fruit is less than 5 letters
    if len(fru) < 5:
      fru_dict["treat"] = "jellybean"
    #. √ else if the fruit has exactly 5 letters, make jelly
    elif len(fru) == 5:
      fru_dict["treat"] = "jelly"
    #. √ else if the fruit starts with a "p", make popsicle:
    elif fru[0] == 'p':
      fru_dict["treat"] = "popsicle"
    #. √ else if the fruit ends with a "y", make jam:
    elif fru[-1] == 'y':
      fru_dict["treat"] = "jam"
    else:
      fru_dict["treat"] = "lollipop"
      #. √ else -- if ALL of the above are false -- make lollipop
    treats.append(fru_dict)

#  save all to new list called treats
print("Fruit Treats:\n")
pp.pprint(treats)
# BONUS: make more useful data structure, list of dict:
# [
#  {"fruit":"apple", "treat":"jelly"},
#  {"fruit":"strawberry", "treat":"jam"},
# ]
```

```python
# Challenge as HW: using in to check for existence of letter in string
vowels = 'aeiou'
# vowels = ['a','e','i','o','u']

# save all to a new treats list of dicts: {"fruit":"apple","treat":"tart"}
treats = []

for fru in fruits:
    fru_dict = {"fruit":fru}
    # if fruit STARTS and ENDS with a VOWEL:
    if fru[0] in vowels and fru[-1] in vowels:
      fru_dict["treat"] = "tart" # make tart
    # elif the fruit STARTS with a vowel:
    elif fru[0] in vowels:
      fru_dict["treat"] = "nectar" # make "nectar"
    # elif the fruit ENDS with a vowel:
    elif fru[-1] in vowels:
      fru_dict["treat"] = "rollup" # make "rollup"
    # else the fruit neither starts nor ends with a vowel
    else:
      fru_dict["treat"] = "jellybean"
    # append the completed dict to list
    treats.append(fru_dict)

# so make a jellybean
pp.pprint(treats)
```

### **String Methods**

- methods include:

  - **str = str.replace(find,repl)** finds char(s) and replaces w other char(s)
  - **str = str.split()** splits a string into a list of words

**BONUS: Scrabble score calculator: iterating a string, letter by letter**
- function takes a user input and returns  its scrabble score
- function loops each word, letter by letter
- each letter's Scrabble value is added to score
- **upper()** string method uppercases string
- **lower()** string method lowercases string

```python
# Scrabble Letter values
# 1 point - A, E, I, O, U, L, N, S, T, R.
#  2 points - D, G.
#  3 points - B, C, M, P.
#  4 points - F, H, V, W, Y.
#  5 points - K.
#  8 points - J, X.
#  10 points - Q, Z.

scrabble_letter_scores = {
  "A": 1, "B": 3, "C": 3, "D": 2, "E": 1, "F": 4, "G": 2, "H": 4, "I": 1, "J": 8,"K": 5, "L": 1, "M": 3, "N": 1, "O": 1, "P": 3, "Q": 10, "R": 1, "S": 1, "T": 1, "U": 1, "V": 4, "W": 4, "X": 8, "Y": 4, "Z": 10
}
```

```python
# Given the dictionary of Scrabble scores
# declare a variable, scrabble_score, set to 0
scrabble_score = 0
vowel_count = 0
VOWELS = 'aeiou'
# make an input for the user to type a word
word = input("Input a word, get its Scrabble score: ")
# loop the inputted word, char by char
for letter in word:
  # look up letter in scrabble_letter_scores dictionary
  letter_score = scrabble_letter_scores[letter.upper()]
  # increment the scrabble score by the score for that letter
  scrabble_score += letter_score
  if letter in VOWELS:
    vowel_count += 1
  # you have to uppercase the letter to match the uppercase keys
  # print the scrabble score at the end
  # bonus:
  # instead of printing just a score, make a dictionary of word + score:
  # examples:

d = {
    "word":word,
    "scrabble_score":scrabble_score,
    "vowel_count":vowel_count
}

print(d)

# { "word": "cat", "scrabble_score": 5 },
# { "word": "zebra", "scrabble_score": 16 }
```
