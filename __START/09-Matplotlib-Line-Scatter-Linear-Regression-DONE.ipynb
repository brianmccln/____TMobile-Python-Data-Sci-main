### **plt.plot(y_values_list) is for line charts**
- takes a list of y-values (such as prices, quantities, units sold, etc.)
- makes a line chart, with values in y-axis
- x-axis is auto incremented as consecutive ints assumed to be a time unit

### **plt.scatter(x_values_list,y_values_list)**
- takes 2 lists of values, one for each axis
- makes a scatter dot chart, with values in both axes
- each dot marks the x,y coordinates of the 2 values
- dots can trend up and to the right for positive correlations
- dots can trend down and to the right for negative correlations

```python
# make a line chart of the stock prices
plt.plot(stock_prices)
plt.title("60 Days Stock Price, Acme Inc.")
plt.xlabel('June-July 2024')
plt.ylabel('US$ Dollars')

plt.ylim(20,120) # set ylim to range from 20-120

# plot the min max prices as scatter dots to highlight these values
# min_max_indices_list: [26, 52]
# min_max_prices_list: [51.5, 94.8]
plt.scatter(min_max_indices_list, min_max_prices_list,
            c=["red","lime"], marker="^", s=150)

# label the high low dots:
plt.text(min_max_indices_list[0]+2,min_max_prices_list[0]-2, 
         str("LOW"),color="red")
plt.text(min_max_indices_list[1]+2,min_max_prices_list[1]-2, 
         str("HIGH"),color="forestgreen")

plt.show() # end the plot w this line always
```

### **m, b = np.polyfit(x_array,y_array,num_slopes)**
- numpy polyfit() method takes 3 args:
-- array of x values
-- array of y values
-- number of slopes (directions of line)
- method returns **m** and **b**, which are for the formula: **y = mx + b** where
-- **m** is the slope of the line and
-- **b** is the y-intercept

```python
Image("../images/regression-best-fit-line.jpg")
```

```python
# make a scatter plot using the two arrays..
# expected result is 14 scattered dots
x = attendance_arr/1000
y = concessions_arr/1000
plt.scatter(x, y, s=5)

plt.ylim(300,1000) # widen / heighten the chart
plt.ylabel("'concessions sales ($1000s)")
# title and label the chart:
plt.title("Concession Sales vs. Attendance Revenue")
plt.xlabel("paid attendance (1000s)")
# plot linear regression line
# attendance,concessions need to be arrays
# slope of a line: y = mx + b
# x, y = the (x,y) point on the line
# m = slope of the line
# b = y-intercept (where the line crosses the y-axis)
# obtain m and b (we only have x, y so far)
# m, b = np.polyfit(x_array,y_array,num_slopes)
m, b = np.polyfit(x,y,1)
# plot the regression in terms of x and y=mx+b
plt.plot(x,m*x+b,color="coral",lw=0.5)
```
