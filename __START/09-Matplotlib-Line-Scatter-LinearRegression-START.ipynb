### **plt.plot(y_values_list) is for line charts**
- takes a list of y-values (such as prices, quantities, units sold, etc.)
- makes a line chart, with values in y-axis
- x-axis is auto incrmented as consecutive ints assumed to be a time unit

### **plt.scatter(x_values_list,y_values_list)**
- takes 2 lists of values, one for each axis
- makes a scatter dot chart, with values in both axes
- each dot marks the x,y coordinates of the 2 values
- dots can trend up and to the right for positive correlations
- dots can trend down and to the right for negative correlations
